### Membership Opeator


```python
## in

## Example with a string
my_list = [10, 20, 30, 40, 50]
print(20 in my_list)  
print(60 in my_list)  

# Example with a string
my_string = "Hello, Python!"
print("Python" in my_string) 
print("Java" in my_string)   
# Example with a tuple
my_tuple = ("apple", "banana", "cherry")
print("banana" in my_tuple) 

# Example with a dictionary (checks for keys)
my_dict = {"name": "Alice", "age": 30}
print("name" in my_dict) 
print("city" in my_dict) 
```

```python
## not in


# Example with a list
my_list = [10, 20, 30, 40, 50]
print(60 not in my_list)  
print(20 not in my_list) 

# Example with a string
my_string = "Hello, Python"
print("Java" not in my_string)    
print("Python" not in my_string) 

# Example with a dictionary (checks for keys)
my_dict = {"name": "Alice", "age": 30}
print("city" not in my_dict)  
print("name" not in my_dict)  
```
