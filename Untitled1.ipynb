```python
## Question 9. 

a1=int(input("enter a1"))
a2=int(input("enter a2"))
b1=int(input("enter b1"))
b2=int(input("enter b2"))
c1=int(input("enter c1"))
c2=int(input("enter c2"))

sum=a1 * b2 - a2 * b1

if sum!=0:
    print("unique solution")
elif a1 * c2 == a2 * c1 and b1 * c2 == b2 * c1:
    print("many solution")
else:
    print("no solution")
```
